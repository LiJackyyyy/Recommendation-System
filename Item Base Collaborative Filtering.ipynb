```python
import pandas as pd
from scipy.spatial.distance import cosine
```

```python
data = pd.read_csv("/content/order.csv")
```

```python
data.head(100)
```

```python
data["quantity"] = 1
```

```python
data = data[["order_id","product_id","quantity"]]
data.head(20)
```

```python
len(pd.unique(data.product_id))
```

```python
dW = data.pivot_table(index="order_id", columns="product_id", values="quantity")
dW
```

```python
dW.iloc[0:2,:]  #觀察
```

```python
dW.iloc[1]  #觀察
```

```python
dW.fillna(0, inplace=True)
dW
```

```python
df1 = dW.copy()
df1
```

```python
df1 = df1.reset_index()
```

```python
df1 = df1.drop("order_id",axis=1)
df1
```

```python
df2 = pd.DataFrame(index=df1.columns,columns = df1.columns)
df2
```

```python
#使用餘弦相似度
for i in range(0,len(df2.columns)) : 
  for j in range(0,len(df2.columns)) :     
    df2.iloc[i,j] = 1-cosine(df1.iloc[:,i],df1.iloc[:,j])
```

```python
df2.head()
```

```python
data_neighbours = pd.DataFrame(index=df2.columns,columns=range(1,5))

for i in range(0,len(df2.columns)):
    data_neighbours.iloc[i,:4] = df2.iloc[0:,i].sort_values(ascending=False)[:4].index
```
